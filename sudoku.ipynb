## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
# Initialise individual cells of the 9x9 sudoku (81 total)
cells = []

# Initialise rows
row = []
rows = []

# Initialise columns
col = []
cols = []

# Initialise 3x3 sub-grids (boxes) of sudoku
box = []
boxes = []

# Define numbers (labels for the columns and digits that fill each cell)
numbers = '123456789'

# Define letters (labels for the rows)
letters = 'ABCDEFGHI'

# Define numbers and letters for boxes
box_numbers = ['123','456','789']
box_letters = ['ABC','DEF','GHI']

### ROW UNIT ###
# Create row & cell variable names
for letter in letters:
    for number in numbers:
        row.append(letter + number)
        cells.append(letter + number)
    rows.append(row)
    row = []   

### COLUMN UNIT ###
# Create column variable names
for number in numbers:
    for letter in letters:
        col.append(letter + number)
    cols.append(col)
    col = []

### BOX UNIT ###
# Create box variable names
i = 0
for letter_group in box_letters:
    for number_group in box_numbers:
        for letter in letter_group:
            for number in number_group:
                box.append(letter + number)
                i += 1
                if i == 9:
                    boxes.append(box)
                    i = 0
                    box = []
                    
# Generates dictionary of "cell:all peers of cell (cells in corresponding row, column and box)"
peers = {}
for box in boxes:
    for cell in box:
        peers[cell] = set(rows[letters.index(cell[0])] + cols[numbers.index(cell[1])] + box) - set([cell])
```

```python
def assign_cell_values(sudoku):
    """
    Converts a sudoku to a dictionary which matches each cell with its corresponding value and returns this.

    Input
        sudoku : 9x9 numpy array
            Empty cells are designated by 0.

    Output
        cell_to_vals : dictionary
            If cell value is already given, value is assigned. If an empty cell, cell value is assigned all possible values 1-9.
    """
    cell_to_vals = {}
    i = 0
    for row in sudoku:
        for cell in row:
            # If cell is non-empty, assign value
            if cell != 0:
                cell_to_vals[cells[i]] = str(cell)
            # If cell is empty, assign all possible values 1-9
            else:
                cell_to_vals[cells[i]] = numbers
            i += 1
            
    return cell_to_vals
```

```python
def find_solved_cells(cell_to_vals):
    """
    Takes a dictionary containing all cell:values and returns a new dictionary containing only those cell:values that have been solved.

    Input
        cell_to_vals : dictionary
            Current state of cells and assigned values.

    Output
        solved_cell_to_vals : dictionary
            Solved cells have only one number in their value.
        
        count : int
            Number of solved cells
    """
    count = 0
    
    solved_cell_to_vals = {}
    # Iterate over all cells and corresponding values in grid
    for cell, number in cell_to_vals.items():
        # If cell is solved (contains one number)
        if len(str(number)) == 1:
            solved_cell_to_vals[cell] = number
            count += 1
            
    return solved_cell_to_vals, count

def find_unsolved_cells(cell_to_vals):
    """
    Takes a dictionary containing all cell:values and returns a new dictionary containing only those cell:values that are unsolved.

    Input
        cell_to_vals : dictionary
            Current state of cells and assigned values.      

    Output
        unsolved_cell_to_num_vals : dictionary
            Unsolved cells have multiple numbers in their value.
    """
    unsolved_cell_to_num_vals = {}
    
    # Iterate over all cells and corresponding values in grid
    for cell, number in cell_to_vals.items():
        # If cell is unsolved (contains more than one number)
        if len(str(number)) > 1:
            unsolved_cell_to_num_vals[cell] = len(str(number))
    
    return unsolved_cell_to_num_vals

def check_solvability(cell_to_vals):
    """
    Takes a dictionary containing all cell:values and checks whether the sudoku is solvable.

    Input
        cell_to_vals : dictionary
            Current state of cells and assigned values.

    Output
        solvable : boolean
            Returns True if sudoku is solvable, else False.
    """
    solvable = True
    
    # Iterate over all cells and corresponding values
    for cell, value in cell_to_vals.items():
        # If a cell has no possible value then sudoku is unsolvable
        if value == '':
            solvable = False
    
    return solvable  
```

```python
def set_value(cell_to_vals, unit):
    """
    Takes a dictionary containing all cell:values and assigns an unsolved cell a value if this value is the only option for that given unit.

    Input
        cell_to_vals : dictionary
            Current state of cells and assigned values.
            
        unit : list
            List of cells in a given row, column or box

    Output
        cell_to_vals : dictionary
            Updated dictionary of cell and assigned values after having assigned singletons.
    """
    # Iterate over each number
    for number in numbers:
        cells_with_number = []
        # Iterate over each cell in unit
        for cell in unit:
            # If number present in cell value of that unit, add to list
            if number in cell_to_vals[cell]:
                cells_with_number.append(cell)
                
        num_cells = len(cells_with_number)

        # If only one cell where value can be assigned, assign cell that value 
        if num_cells == 1:
            singleton_cell = cells_with_number.pop()
            cell_to_vals[singleton_cell] = number
        else:
            cells_with_number = []
    
    return cell_to_vals   

def constraint_propagation(cell_to_vals):
    """
    Takes a dictionary containing all cell:values and performs two methods of constraint propagation.

    Input
        cell_to_vals : dictionary
            Current state of cells and assigned values.

    Output
        cell_to_vals : dictionary
            Updated dictionary of cell and assigned values after performing constraint propagation.
        
        progress : boolean
            Returns True if constraint propagation has solved cells, else returns False.
    """
    solved_cell_to_vals, num_old_solved_cells = find_solved_cells(cell_to_vals)
    
    """Method 1 - removes potential values from peers' domain of a given solved cell""" 
    for cell, number in solved_cell_to_vals.items():
        peers_of_cell = peers[cell]
        
        # For each peer of a solved cell, remove number from their possible values
        for peer in peers_of_cell:
            cell_to_vals[peer] = cell_to_vals[peer].replace(number, '')  

    """Method 2 - assigns an unsolved cell a value if this value is the only option for that given row, column or box"""            
    for row in rows:
        cell_to_vals = set_value(cell_to_vals, row)
        
    for col in cols:
        cell_to_vals = set_value(cell_to_vals, col)
        
    for box in boxes:
        cell_to_vals = set_value(cell_to_vals, box)
                                   
    _, num_new_solved_cells = find_solved_cells(cell_to_vals)
    
    # Checks if number of solved cells has changed
    if num_old_solved_cells != num_new_solved_cells:
        progress = True
    else:
        progress = False
                            
    return cell_to_vals, progress

def reduce_search_space(cell_to_vals):
    """
    Takes a dictionary containing all cell:values and reduces the search space.

    Input
        cell_to_vals : dictionary
            Current state of cells and assigned values.

    Output
        cell_to_vals : dictionary
            Updated dictionary of cell and assigned values.
    """
    while True:
        # Run constraint propagation methods
        cell_to_vals, progress = constraint_propagation(cell_to_vals)
        
        # If constraint propagation no longer solving any cells, exit
        if not progress:
            break    
            
    return cell_to_vals 
```

```python
def depth_first_search(cell_to_vals):
    """
    Takes a dictionary containing all cell:values and executes depth first search.

    Input
        cell_to_vals : dictionary
            Current state of cells and assigned values.

    Output
        trial : dictionary
            Updated dictionary of cell and assigned values having performed depth first search.
        OR
        
        False : boolean
            No solution.
    """
    # Find all unsolved cells
    unsolved_cell_to_num_vals = find_unsolved_cells(cell_to_vals)
    
    # Select the cell with the smallest number of possible values
    potential_cell = min(unsolved_cell_to_num_vals, key=unsolved_cell_to_num_vals.get)
    
    # Obtain values for given cell
    possible_values = cell_to_vals[potential_cell]
    
    for val in possible_values:
        # Make a copy of current sudoku
        updated_sudoku = cell_to_vals.copy()
        # Assign a possible value to the cell
        updated_sudoku[potential_cell] = val
        # Attempt to find a solution with updated sudoku
        trial = find_solution(updated_sudoku)

        if trial is not False:
            return trial
        
    return False # After recursive search and constraint propagation, solution cannot be found
```

```python
def find_solution(cell_to_vals):
    """
    Takes a dictionary containing all cell:values and runs constraint propagation and depth first search to find a solution.

    Input
        cell_to_vals : dictionary
            Current state of cells and assigned values.

    Output
        cell_to_vals : dictionary
            Updated dictionary of cell and assigned values having performed constraint propagation and depth first search.
        OR
        
        False : boolean
            No solution.
    """
    cell_to_vals = reduce_search_space(cell_to_vals)
    
    is_solvable = check_solvability(cell_to_vals)
        
    # If sudoku is unsolvable or no solution
    if not is_solvable or cell_to_vals is False:
        return False
    
    if cell_to_vals is not False:
        _, count = find_solved_cells(cell_to_vals)
        # If all cells solved
        if count == 81:
            return cell_to_vals 
    
    cell_to_vals = depth_first_search(cell_to_vals)
    
    return cell_to_vals

def display_solved_sudoku(solution):
    """
    Converts a dictionary which matches each solved cell with its corresponding value, to an array which displays the sudoku and returns this.

    Input
        solution : dictionary
            All solved cells matched with their corresponding number. 

    Output
        sudoku : 9x9 numpy array
            Solved sudoku.
    """
    solved_sudoku = []
    i = 0

    solved_list = list(solution.values())
    while i <= 72:
        current_row = solved_list[i:i+9]
        for j in range(len(current_row)):
            current_row[j] = int(current_row[j])
        solved_sudoku.append(current_row)
        i += 9   
        
    solved_sudoku = np.array(solved_sudoku)
    
    return solved_sudoku

def sudoku_solver(sudoku):
    """
    Solves a Sudoku puzzle and returns its unique solution.

    Input
        sudoku : 9x9 numpy array
            Empty cells are designated by 0.

    Output
        9x9 numpy array of integers
            It contains the solution, if there is one. If there is no solution, all array entries should be -1.
    """
    cell_to_vals = assign_cell_values(sudoku)
    solution = find_solution(cell_to_vals)

    if solution is not False:
        solved_sudoku = display_solved_sudoku(solution)
    else:
        solved_sudoku = (np.ones((9,9))*-1).astype(np.int)
    
    return solved_sudoku
```
